```python
# 把所有用户id映射为数字
fr = open('/home/data/TalkingData_location_Tqinghua.csv','r')
from time import time 
t0 = time()
idtable = {}
data_table = {}  # 将数据存在另外一个字典里面，方便按照用户id存到文件
index = 1
count = 0
for line in fr.readlines():
    item = [k.strip() for k in line.split('\t')]
    if len(item) != 5:
        continue
    if item[4] != 'people':
        continue
    if int(item[1]) > 5000:
        continue
    count = count + 1  
    if count <= 80000000: 
        if item[0] in idtable:      # 按照一样的编号
            pass         
        else:
                idtable[item[0]] = str(index)
                data_table[idtable[item[0]]] = []
                index = index + 1
                continue
    if count > 80000000: 
        if item[0] in idtable:
            ind = idtable[item[0]]
            data_table[ind].extend(item[1:4])
        else:
                idtable[item[0]] = str(index)
                data_table[idtable[item[0]]] = item[1:4]
                index = index + 1  
t1 = time()
fr.close()
print(index)
print(t1-t0)
```

```python
# 将0-80000000的数据写入文件data_lessthan_10e8.csv中
# 将80000001-end的数据写入文件data_morethan_10e8.csv中
fw = open('/home/locus/data_morethan_10e8.csv','w')
t1 = time()
for i in range(1, index):
    num = int(len(data_table[str(i)])/3)  
    if num == 0:
        continue
    for j in range(1, num+1):
        fw.write(str(i)+','+data_table[str(i)][(j-1)*3]+','+data_table[str(i)][(j-1)*3+1]+','+data_table[str(i)][(j-1)*3+2])   
        fw.write('\n')
fw.close()
t2 = time()
print(t2 - t1)

```

```python
import time as time
fr = open('/home/locus/databyorder.csv','r')
fw = open('/home/data/cleandata_by_order.csv','w')
t1 = time.time()
count1 = 0
count2 = 0
last = 0
id_num = 0
result = []
place = []
StaticUser = []     #位置不变的用户
for ed in fr.readlines():
    item =[k.strip() for k in ed.split(',')]    
    if last != int(item[0]):
        id_num = id_num + 1
        if last == 0:
            pass
        elif len(set(place)) == 1:
            StaticUser.append(int(item[0]))
            result = []
            place = []
        elif len(set(place)) > 1:
            g = int(len(result)/4)
            for c in range(0, g):
                fw.write(result[c*4]+','+result[c*4+1]+','+result[c*4+2]+','+result[c*4+3])
                fw.write('\n')
                count2 += 1
            result = []
            place = []        
    result.extend([item[0],item[1],item[2],item[3]])
    place.append((item[2],item[3]))         
    last = int(item[0])
    count1 += 1
fr.close()
fw.close()
```
